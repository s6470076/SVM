# SVM
- Support Vector Machine
- 선형SVM과 비선형SVM이 있다
- 분류수행, 회귀수행
- 비선형적이고 고차원 공간에서도 탁월한 성능을 보임
- 디시전트리는 곡선으로 쪼개지 못 하지만 서포터 벡터 머신은 가능
- 단점 : 알고리즘이 수학적 이론이라 어려운 모델
- 장점 : 현실세계의 모델에서 복잡란 문제를 푸는 능력이 좋은 모델
- 클레스를 적당하게 분류할 수 있는 적당한 선분을 찾는 알고리즘
- support vector machine은 margin을 최대로 하는 서포트 벡터를 찾는 것
- SVM은 멀티클래스도 가능함

# 비선형SVM
- 비선형적인 문제를 선형적인 문제로 바꾸려면 매핑함수 필요
- Decision Tree는 엔트로피에 의해서 데이터를 선형으로 나눔
    - 비선형으로 나누는 것이 힘듦
    - 데이터가 비선형인 경우에는 잘 맞지 않음
    - 이것을 해결할 수 있는 방법이 비선형 SVM이다
### 매핑함수를 사용하면?
    - 데이터 = 선형적인 형태로 변형 
    - 축 = 매우 복잡한 형태로 변형 => 축을 설명하기 더 곤란해지는 상황 발생
